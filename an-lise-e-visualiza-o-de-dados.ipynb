### 3° Passo: Preparando os Dados
* Retirando colunas que não serão utilizadas
* Identificando duplicidade
* Renomeando colunas
* Criando recursos

```python
df_IDH = df_IDH[['UF','HDI 1991','HDI 2000','HDI 2010','HDI 2017']]
```

```python
df_estados = df_estados.rename(columns={'State':'Estado','Region':'Regiao', 'Population':'População',
                                        'Demographic Density':'Densidade Demografica', 'Cities count':'Numero de cidades',
                                        'GDP':'PIB', 'GDP rate':'PIB per Capita', 'Poverty':'Indice de Pobreza'})
```

```python
df_IDH = df_IDH.rename(columns={'HDI 1991':'IDH 1991','HDI 2000':'IDH 2000','HDI 2010':'IDH 2010','HDI 2017':'IDH 2017'})
```

```python
df_estados.loc[df_estados['Regiao'].str.contains('Northeast'),'Regiao'] = 'Nordeste'
df_estados.loc[df_estados['Regiao'].str.contains('Southeast'),'Regiao'] = 'Sudeste'
df_estados.loc[df_estados['Regiao'].str.contains('North'),'Regiao'] = 'Norte'
df_estados.loc[df_estados['Regiao'].str.contains('South'),'Regiao'] = 'Sul'
df_estados.loc[df_estados['Regiao'].str.contains('Center-west'),'Regiao'] = 'Centro-Oeste'
```

```python
maior_dens_dem = df.sort_values('Densidade Demografica', ascending = False).head()
menor_dens_dem = df.sort_values('Densidade Demografica', ascending = False).tail()

fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=False)

sns.barplot(ax = axes[0], x = maior_dens_dem['UF'], y = maior_dens_dem['Densidade Demografica']).set(xlabel = 'Estados', ylabel = 'Densidade Demográfica')
axes[0].set_title('Maiores Densidades Demográficas')

sns.barplot(ax = axes[1], x=menor_dens_dem['UF'], y=menor_dens_dem['Densidade Demografica'],  palette = 'Set2').set(xlabel='Estados', ylabel='Densidade Demográfica')
axes[1].set_title('Menores Densidades Demográficas')
plt.show()
```

* O Distrito Federal é um outlier por ter somente uma cidade apesar de estar contando como um estado.
* A diferença na densidade demográfica entre os maiores e menores estados é muito alta tendo RJ com valor em torno de 400 e o AM 2.5

```python
maior_ind_pob = df.sort_values('Indice de Pobreza',ascending=False).head()
menor_ind_pob = df.sort_values('Indice de Pobreza',ascending=False).tail()

fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=False)


sns.barplot(ax=axes[0], x=maior_ind_pob['UF'], y=maior_ind_pob['Indice de Pobreza']).set(xlabel='Estados', ylabel='Índices de Pobreza')
axes[0].set_title('Maiores Índice de Pobreza')

sns.barplot(ax=axes[1], x=menor_ind_pob['UF'], y=menor_ind_pob['Indice de Pobreza'],  palette = 'Set2').set(xlabel='Estados', ylabel='Índices de Pobreza')
axes[1].set_title('Menores Índice de Pobreza')

plt.show()
```

O Índice de pobreza é calculado com o número de pessoas que recebem menos que U$ 5,5 por dia dividido pela população e divido por 100 para gerar esse índice que se multiplicado por 100 encontrariamos a porcentagem da população

### 5° Passo: Correlações 

```python
corr = df.corr(method ='pearson')
```

```python
sns.heatmap(corr, cmap="rocket_r", annot=False)
plt.show()
```

Neste gráfico foi utilizado o coeficiente de correlação de Pearson que mede o grau da correlação entre duas variáveis e assume apenas valores entre -1 e 1 para implicar a correlação entre essas váriaveis.
Com basé nesse gráfico podemos ver algumas correlações:
* PIB e PIB per capita com IDH
* Índice de Pobreza com Latitude

```python
df_city_area = df_region.copy()
df_city_area['Area/Num Cid'] = df_region['Area']/df_region['Numero de cidades'].sort_values()
df_city_area = df_city_area.reset_index().sort_values('Area/Num Cid',ascending=False)
```

```python
fig, axes = plt.subplots(figsize=(8, 6))
sns.barplot(ax=axes,data=df_city_area, x="Regiao", y="Area/Num Cid").set(xlabel='Região', ylabel='Aréa / Número de Cidades')

plt.show()
```

Foi feito uma cáculo para termos uma ideia da "média" da área de cada cidade e foi exibido por região, onde é possível identificar que o Norte do país tem um número muito maior de área por número de cidade 

```python
df_cres = df.copy()
df_cres['Crescimento IDH'] = df_cres['IDH 2017'] - df_cres['IDH 1991']
```

```python
maior_cres_IDH = df_cres.sort_values('Crescimento IDH',ascending=False).head()
menor_cres_IDH = df_cres.sort_values('Crescimento IDH',ascending=False).tail()

fig, axes = plt.subplots(1, 2, figsize=(12, 6),sharey=True)

sns.barplot(ax=axes[0], x = maior_cres_IDH['UF'], y = maior_cres_IDH['Crescimento IDH']).set(xlabel = 'Estados', ylabel = 'Crescimento do IDH')
axes[0].set_title('Maiores Crescimentos')

sns.barplot(ax=axes[1], x = menor_cres_IDH['UF'], y = menor_cres_IDH['Crescimento IDH'], palette = 'Paired').set(xlabel = 'Estados', ylabel = 'Crescimento do IDH')
axes[1].set_title('Menores Crescimentos')

plt.show()
```

Para encontrar o valor desse eixo Y foi feito uma subtração do valor do IDH de 2017 pelo de 1991.
* Foi possíve ver que todos os estados tiverem um valor positivo.

```python
df_grupo_dens_dem = df.copy()
df_grupo_dens_dem['Grupo Densidade Demografica'] = '0-10'
df_grupo_dens_dem.loc[(df_grupo_dens_dem['Densidade Demografica'] > 10) & (df_grupo_dens_dem['Densidade Demografica'] <= 60) , 'Grupo Densidade Demografica'] = '011-60'
df_grupo_dens_dem.loc[(df_grupo_dens_dem['Densidade Demografica'] > 60) & (df_grupo_dens_dem['Densidade Demografica'] <= 100) , 'Grupo Densidade Demografica'] = '061-100'
df_grupo_dens_dem.loc[(df_grupo_dens_dem['Densidade Demografica'] > 100) & (df_grupo_dens_dem['Densidade Demografica'] <= 200) , 'Grupo Densidade Demografica'] = '101-200'
df_grupo_dens_dem.loc[df_grupo_dens_dem['Densidade Demografica'] > 200, 'Grupo Densidade Demografica'] = '200+'

df_grupo_dens_dem = df_grupo_dens_dem.sort_values('Grupo Densidade Demografica')
```

```python
mask1= df_grupo_dens_dem["Longitude"] != 0
mask2 = df_grupo_dens_dem["Latitude"] !=0 
mask3 = df_grupo_dens_dem["Area"] != 0

x = df_grupo_dens_dem[mask1&mask2]["Longitude"]
y = df_grupo_dens_dem[mask1&mask2]["Latitude"]
z = df_grupo_dens_dem[mask1&mask2]["Area"]

color = ['#f9f3a9','#efdf88','#f3a823','#ed6f2a','#c73d1d']
sns.set(rc={'figure.figsize':(12,12)})
sns.set_style('white')
sns.scatterplot( x="Longitude", y="Latitude", data=df_grupo_dens_dem[mask1&mask2&mask3], 
           palette='rocket_r',hue='Grupo Densidade Demografica',style='Regiao',size=z\
                ,sizes=(30, 250),linewidth=0.3,edgecolor='black')
plt.show()
```
